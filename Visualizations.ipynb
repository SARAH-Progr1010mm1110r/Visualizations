```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.plotting.register_matplotlib_converters()
```

```python
Raisin = pd.read_csv('Raisin_Dataset.csv')
```

```python
Raisin.head()
```

```python
# Line chart showing the number of pixels in image of raisin
plt.figure(figsize=(8, 6))
sns.lineplot(data=Raisin['Area'])
plt.title("number of pixels in image of raisin.")
```

```python
# Bar chart showing Average of Class of Rainsin
plt.figure(figsize=(8, 6))
sns.barplot(x=Raisin['Class'], y=Raisin.index)
plt.title("Average of Class of Rainsin")
```

```python
# Scatter plot showing the relationship between 'Area' and 'ConvexArea'
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Raisin['MajorAxisLength'], y=Raisin['MinorAxisLength'],hue=Raisin['Class'], alpha=0.5)
plt.title(" relationship between Area and ConvexArea")
```

```python
# Pie Chart showing how many Class of raisin
plt.figure(figsize=(8, 6))
Raisin['Class'].value_counts().head(10).plot.pie()
plt.gca().set_aspect('equal')
plt.title("show how many Class of raisin")
```

```python
# Histogram show number of pixels in image of raisin
plt.figure(figsize=(8, 6))
sns.displot(Raisin, x="Area")
plt.title("show number of pixels in image of raisin")
```

```python
# Boxplot show number of pixels in image of raisin
plt.figure(figsize=(8, 6))
sns.boxplot(x=Raisin["Area"])
plt.title("number of pixels in image of raisin")
```
